```python
#import pandas library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('ICC Men T20.csv')
df.head()# Head method is used to display the top 5 rows of the dataset
```

```python
# Convert date column to datetime format for better analysis
df["Date of Match"] = pd.to_datetime(df["Date of Match"], format="%d-%m-%Y")
```

```python
# Top 5 teams with the most wins
top_winners = df["Winners"].value_counts().head(5)
plt.figure(figsize=(7,4))
sns.barplot(x=top_winners.index, y=top_winners.values, palette="viridis")
plt.title("Top 5 Teams with Most Wins")
plt.xlabel("Team")
plt.ylabel("Wins")
plt.xticks(rotation=45)
plt.show()
```

```python
#Toss Decision Analysis
plt.figure(figsize=(6,4))
sns.countplot(x=df["Toss Decision"],palette="coolwarm")
plt.title("Toss Decision Distribution")
plt.xlabel("Decision")
plt.ylabel("Count")
plt.show()
```

```python
# Super Over Matches Count
super_over_count = df["Super Over Match"].value_counts()
plt.figure(figsize=(6, 4))
sns.barplot(x=super_over_count.index, y=super_over_count.values, palette="pastel")
plt.title("Super Over Matches Count")
plt.xlabel("Super Over")
plt.ylabel("Count")
plt.show()
```

```python
# Top 5 teams with the most wins
VenueCity = df["Venue City"].value_counts()
plt.figure(figsize=(7,4))
sns.barplot(x=VenueCity.values, y=VenueCity.index, palette="viridis")
plt.title("cities in major matches are held")
plt.xlabel("No of matches held in the city")
plt.ylabel("Cities")
plt.show()
```

```python
# Convert the "Date of Match" column to datetime format
df["Date of Match"] = pd.to_datetime(df["Date of Match"], format="%d-%m-%Y")
# Sort data by date
df = df.sort_values(by="Date of Match")
# Aggregate total scores per day
df["Total Runs"] = df["First Innings Score"] + df["Second Innings Score"]
time_series = df.groupby("Date of Match")["Total Runs"].sum()
```

```python
# Plot time series of total match scores
plt.figure(figsize=(12, 6))
plt.plot(time_series.index, time_series.values, marker="o", linestyle="-", color="b", label="Total Runs per Matchday")
plt.xlabel("Date")
plt.ylabel("Total Runs")
plt.title("Time Series Analysis of Total Runs in ICC T20 World Cup Matches")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
#set Plot Style 
sns.set(style="whitegrid")
```

```python
#Create subplots for numerical data distributions
fig,axes=plt.subplots(2,2, figsize=(12,10))

#Histogram of first innings score
sns.histplot(df["First Innings Score"], bins=10, kde=True, ax=axes[0, 0])
axes[0, 0].set_title("Distribution of First Innings Scores")

# Histogram of Second Innings Score
sns.histplot(df["Second Innings Score"], bins=10, kde=True, ax=axes[0, 1])
axes[0, 1].set_title("Distribution of Second Innings Scores")

# Winning Margin Distribution
sns.histplot(df["Winning Margin"], bins=10, kde=True, ax=axes[1, 0])
axes[1, 0].set_title("Winning Margin Distribution")

# Correlation heatmap
corr_matrix = df.select_dtypes(include=['int64']).corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", ax=axes[1, 1])
axes[1, 1].set_title("Correlation Matrix")
# Visualize the data
plt.tight_layout()
plt.show()
```
